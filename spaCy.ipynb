```python
import spacy
from spacy import displacy
```

```python
# Importieren der Deutschen und Englischen Sprachannotationen
# Das sind jeweils die kleinsten Versionen ~20-30mb
# gro√üe Versionen reichen bis an die 800mb

nlp = spacy.load("en_core_web_sm")
nlp_de = spacy.load("de_core_news_sm")
```

```python
text = ("Hello there, my name is John. I am a huge Star Wars fan and also like Ferrari.")

doc = nlp(text)
```

```python
print('-----------------------------------')
for token in doc:
    print(token.text, token.pos_, token.dep_)
```

```python
# Nun folgt ein deutscher Text

text_de = "Dies ist ein deutscher Text. Ich kaufe immer bei Apple ein und mache viel Kohle."

doc_de = nlp_de(text_de)
```

```python
# spaCy kann auch Eigennamen (named entities) sehr leicht erkennen
for ent in doc_de.ents:
    print(ent.text, ent.start_char, ent.end_char, ent.label_)
```

## displaCy kann die Dependenzstrukturen dann ganz leicht sichtbar machen:

```python
displacy.render(doc, style="dep", jupyter=True, options={'distance': 140})
```

```python
#Damit wir einzelne Saetze rendern koennen, machen wir aus dem Dokument eine Liste:
sentences = list(doc.sents)
displacy.render(sentences[0], style="dep", jupyter=True, options={'distance': 140})
```
